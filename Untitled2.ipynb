```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load dataset
file_path = "gameandgrade.csv"  
# Change this if needed
df = pd.read_csv(file_path)
# Remove any extra dots and non-numeric characters
df['Grade'] = df['Grade'].astype(str).str.replace(r'[^0-9.]', '', regex=True)  # Keep only numbers and dots

# Handle cases where there are multiple dots (e.g., '92..00')
df['Grade'] = df['Grade'].str.replace(r'\.{2,}', '.', regex=True)  # Replace multiple dots with a single dot

# Convert to float
df['Grade'] = pd.to_numeric(df['Grade'], errors='coerce')  # Convert to float, set errors to NaN if invalid

# Check if there are still any NaN values after conversion
print(df['Grade'].isna().sum(), "invalid values converted to NaN")
print(df['Grade'].dtype)  # Should print float64 or int64
```

```python
# Basic info and summary
print("Dataset Information:\n")
df.info()
print("\nSummary Statistics:\n", df.describe())
```

```python
missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)

# In this dataset, if there are missing values, we perform a simple imputation or drop rows based on the extent of missingness.
if missing_values.sum() > 0:
    # Given the small dataset size in many Kaggle competitions, dropping may be acceptable
    df = df.dropna()
    print('Dropped rows with missing values')

# Check data types
print('\nData Types:')
print(df.dtypes)

# Convert the Grade column to a categorical type (if it is not already)
df['Grade'] = df['Grade'].astype('category')

print('\nUnique grades:', df['Grade'].unique())
```

```python
# Histogram of numeric columns
numeric_columns = ['Sex', 'School Code', 'Playing Years', 'Playing Often', 'Playing Hours', 'Playing Games', 'Parent Revenue', 'Father Education', 'Mother Education']
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
    

# Count plot for the Grade distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='Grade', data=df, palette='viridis')
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Group by '' and calculate the mean of 'Grade'
mother_edu_avg_grade = df.groupby('Mother Education', as_index=False)['Grade'].mean()

# Plot the bar graph
plt.figure(figsize=(8, 4))
sns.barplot(x='Mother Education', y='Grade', data=mother_edu_avg_grade, palette='Purples')
plt.title("Mean Grade vs Mother's Education")
plt.xlabel("Mother's Education Level")
plt.ylabel("Mean Grade")
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10, 8))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Numeric Features Correlation Heatmap')
    plt.tight_layout()
    plt.show()
else:
    print('Not enough numeric features to create a correlation heatmap.')

"""
following heat map shows the interdependance of variables on each other
the gender and school code are categorical variables so they do not have any corellation with other variables
"""
```
